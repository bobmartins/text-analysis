```python
import pandas as pd
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
text = pd.read_parquet("../data/tour-text.parquet")
```

```python
text
```

```python
vectors = pd.read_parquet("data/tour-vectors.parquet")
```

```python
text['documents'] = text.apply(lambda doc: TaggedDocument(doc.tokens, [str(doc.tour_id), str(doc.main_topic_id+1), doc.poi_type_name]), axis=1)
```

```python
max_epochs = 500
vec_size = 300
alpha = 0.05
NUM_WORKERS = 32

model = Doc2Vec(vector_size=vec_size, alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1,
                dbow_words=1,
                workers=NUM_WORKERS)

model.build_vocab(text.documents.values)
```

```python
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(text.documents.values,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    # decrease the learning rate
    model.alpha *= 0.995
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
```

```python
model.save('d2v_tourtext_3.pck')
```

```python
model.docvecs.most_similar(positive=[model.docvecs.index2entity.index('35191')])
```

```python
from tsnecuda import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from numpy.random import choice
```

```python
plot_data = choice(text.shape[0], 5000, False)
```

```python
tsne_model = TSNE(n_components=2, verbose = 1, perplexity=50.0, learning_rate=200.0, n_iter=5000)
tsne_articles_2D = tsne_model.fit_transform(model.docvecs.vectors_docs[plot_data])
```

```python
import matplotlib
import matplotlib.cm as cm

types = {k: v for (v, k) in enumerate(text.category_type[plot_data].unique())}
norm = matplotlib.colors.Normalize(vmin=0., vmax=len(types), clip=True)
mapper = cm.ScalarMappable(norm=norm, cmap=cm.viridis)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(tsne_articles_2D[:,0], tsne_articles_2D[:,1], 
           c=[mapper.to_rgba(types[x]) for x in text.category_type[plot_data]])
plt.show()
```
