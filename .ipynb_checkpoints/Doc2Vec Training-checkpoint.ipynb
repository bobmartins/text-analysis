```python
import pandas as pd
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
text = pd.read_parquet("data/tour-text.parquet")
```

```python
text
```

```python
vectors = pd.read_parquet("data/tour-vectors.parquet")
```

```python
text['documents'] = text.apply(lambda doc: TaggedDocument(doc.tokens, [str(doc.tour_id), str(doc.main_topic_id+1), doc.poi_type_name]), axis=1)
model = Doc2Vec(text.documents, vector_size=50, window=5, min_count=50, workers=4, dm=1, alpha=0.025, min_alpha=0.0025, sample=1e-5, epochs=1)
```

```python
model.docvecs.most_similar(positive=[model.docvecs.index2entity.index('28147')])
```
